```python
import pytextrank as ptr
import networkx as nx
import pylab as plt

def extractKeyword(inputJsonFile):
    
    tempStageOneFile = "stageOne.json"
    tempStageTwoFile = "stageTwo.json"
    
    with open(tempStageOneFile, 'w') as f:
        for tempGraph in ptr.parse_doc(ptr.json_iter(inputJsonFile)):
            f.write("%s\n" % ptr.pretty_print(tempGraph._asdict()))


    graph, ranks = ptr.text_rank(tempStageOneFile)
    ptr.render_ranks(graph, ranks)
    
    with open(tempStageTwoFile, 'w') as f:
        for line in ptr.normalize_key_phrases(tempStageOneFile, ranks):
            f.write("%s\n" % ptr.pretty_print(line._asdict()))

    phrases = ", ".join(set([p for p in ptr.limit_keyphrases(tempStageTwoFile, phrase_limit=5)]))

    nx.draw(graph, with_labels=True) 
    plt.show()

    return phrases
```

```python
inputJsonFile = "kpe-text-one.json"
phrases = extractKeyword(inputJsonFile)
print(phrases)
```
