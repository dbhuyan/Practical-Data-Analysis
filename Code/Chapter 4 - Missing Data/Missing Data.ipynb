# Missing Data - A view from above

Data Source - https://data.ny.gov/Transparency/Real-Property-Transactions-of-State-Authorities/t7uh-5ac8

```python
import pandas as pd
import missingno as mn
%matplotlib inline
df = pd.read_csv('Real_Property_Transactions_of_State_Authorities.csv',
                 encoding = "ISO-8859-1")
```

### Matrix View

```python
mn.matrix(df)
```

### Bar Chart View

```python
mn.bar(df)
```

### Heatmap View

```python
mn.heatmap(df)
```

### Dendrogram View

```python
mn.dendrogram(df)
```
